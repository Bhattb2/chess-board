```python
class ChessBoard:

    white = (1, 1, 1)
    red = (1, .2, 0)
    blue = (.1, .1, 1)

    def __init__(self):
        self._grid = np.zeros((8, 8, 3))
        self.red_row = None
        self.red_col = None
        self.blue_row = None
        self.blue_col = None

    def render(self):
        # This creates the chess board.
        self._grid = np.zeros((8, 8, 3))
        for i in range(0, 8, 2):
            self._grid[i, 0: 8: 2] = [1, 1, 1]
            self._grid[i+1, 1: 8: 2] = [1, 1, 1]
        if self.red_row !=None or self.red_col !=None:
            self._grid[self.red_row,self.red_col] = (1, 0.2, 0)
        if self.blue_row !=None or self.blue_col !=None:
            self._grid[self.blue_row, self.blue_col] = (.1, .1, 1)

        plt.imshow(self._grid)

    def color_cell(self, row, col, color):
        self._grid[row, col] = color
    
    def add_red(self, row, col):
        self.red_row = row
        self.red_col = col
        self._grid[row, col] = (1, .2, 0)
        self.render()

    def add_blue(self,row,col):
        self.blue_row = row
        self.blue_col = col
        self._grid[row, col] = (0.1, .1, 1)
        self.render()

    def is_diagonal(self):
        rows = abs(self.red_row - self.blue_row)
        cols = abs(self.red_row - self.blue_col)
        return rows == cols

    def is_rows(self):
        rows = abs(self.red_row - self.blue_row)
        return rows == 0

    def is_cols(self):
        cols = abs(self.red_col - self.blue_col)
        return cols == 0

    def is_under_attack(self):
        #check if the red and the blue color are under attack
        return self.is_diagonal() or self.is_rows() or self.is_cols()
    
    
```

```python
chess = ChessBoard()
```

```python
chess.render()
```

```python
chess.add_red(5, 5)
```

```python
chess.add_blue(3, 4) 
```

```python
chess.add_blue(3,3)
chess.add_red(1,2)
chess.render()
actual= chess.is_under_attack()
# assert actual == False
print(actual)
```

```python
chess.add_blue(1,7)
chess.add_red(1,2)
chess.render()
actual= chess.is_under_attack()
assert actual == True
```
